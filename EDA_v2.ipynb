```python
# Ruta de la carpeta con archivos
path_data = '../ENACOM_Data_Analyst/ENACOM_data/'

# Obtener un listado con todos los archivos de la carpeta
lista_archivos_cvs = os.listdir(path_data)
lista_archivos_cvs
```

```python
len(lista_archivos_cvs)
```

```python
# Con esta función genero una tubla separando nombre y extension
os.path.splitext(lista_archivos_cvs[0])
```

```python
# Puedo seleccionar solo el nombre para guardarlo para nombrar los df
os.path.splitext(lista_archivos_cvs[0])[0]
```

```python
path_data + lista_archivos_cvs[0]
```

```python
# Realizo un diccionario de dataframes
dic_df_archivos = {}
for archivo in lista_archivos_cvs:
    df_temp = pd.read_csv(path_data + archivo, index_col= 0)
    nombre = os.path.splitext(archivo)[0]
    dic_df_archivos[nombre] = df_temp
```

```python
# Retiro el contenido del diccionario como Dataframes independientes
for df_name, df in dic_df_archivos.items():
    globals()[df_name] = df
```

## Revision de archivos

```python
# Veamos la cantidad de información de cada archivo
print(f'----------------------------------------------------')
print(f'|{"Nombre":35}|{"Filas":5}|{"Columnas":8}|')

for df_name, df in dic_df_archivos.items():
    print(f'|-----------------------------------|-----|--------|')
    print(f'|{df_name:35}|{df.shape[0]:5}|{df.shape[1]:8}|')

print(f'----------------------------------------------------')
```
